```python
import numpy as np
import pandas as pd
from yellowbrick.features import PCA
from yellowbrick.features import Manifold
```

```python
abobora = pd.read_excel('Pumpkin_Seeds_Dataset.xlsx')
abobora.head()
```

```python
label= np.where(abobora['Class']=='Çerçevelik',0,1)
```

```python
visualizer = PCA(scale=True, classes=abobora['Class'].unique())
visualizer.fit_transform(abobora.drop('Class',axis=1), label)
visualizer.show()
```

```python
visualizer = PCA(scale=True, classes=abobora['Class'].unique(),projection=3)
visualizer.fit_transform(abobora.drop('Class',axis=1), label)
visualizer.show()
```

```python
visualizer = PCA(scale=True, classes=abobora['Class'].unique(),proj_features=True)
visualizer.fit_transform(abobora.drop('Class',axis=1), label)
visualizer.show()
```
